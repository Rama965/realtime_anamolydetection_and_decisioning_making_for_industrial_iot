```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
# Modelling
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, Ridge,Lasso
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
data.columns
```

```python
cat_cols = ['Machine_ID', 'Machine_Type', 'Installation_Year', 'Operational_Hours',
       'Temperature_C', 'Vibration_mms', 'Sound_dB', 'Oil_Level_pct',
       'Coolant_Level_pct', 'Power_Consumption_kW',
       'Last_Maintenance_Days_Ago', 'Maintenance_History_Count',
       'Failure_History_Count', 'AI_Supervision', 'Error_Codes_Last_30_Days',
       'Remaining_Useful_Life_days', 'Failure_Within_7_Days',
       'AI_Override_Events']
for x in cat_cols:
    if data[x].dtype == 'object':
        data[x] = LabelEncoder().fit_transform(data[x].astype(str))
    elif data[x].dtype == 'bool':
        data[x] = LabelEncoder().fit_transform(data[x].astype(str))
```

```python
data
```

```python
X = data.drop(['Machine_ID','Machine_Type','Installation_Year','Operational_Hours','Failure_Within_7_Days'],axis=1)
X
```

```python
X.columns
```

```python
y = data['Failure_Within_7_Days']
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
scaled_data
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_data,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
import xgboost as xgb
model = xgb.XGBClassifier(subsamples=0.5,object='binary:logistic')
```

```python
params = {'n_estimators':[10,25,50,100],
         'max_depth':[4,8,12,16],
         'learning_rate':[0.1,0.01,0.001,1.0]}
```

```python
grd = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy')
grd.fit(x_train,y_train)
```

```python
grd.best_params_
```

```python
final_model = grd.best_estimator_
final_model
```

```python
yhat_train = final_model.predict(x_train)
yhat_train
```

```python
print(classification_report(y_train,yhat_train))
cm = confusion_matrix(y_train,yhat_train)
sns.heatmap(cm,annot=True,fmt='d')
```

```python
yhat_test = final_model.predict(x_test)
yhat_test
```

```python
print(classification_report(y_test,yhat_test))
cm = confusion_matrix(y_test,yhat_test)
sns.heatmap(cm,annot=True,fmt='d')
```
